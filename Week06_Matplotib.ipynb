```python
#@author : Kevin Saputra 2473011
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

```python
#Box Plot
#Sample dataset
np.random.seed(2)
data = np.random.normal(loc=100, scale=10, size=200)
data
```

```python
#Simple Box plot
## DEPRECATED
plt.boxplot(data, labels=['Data'])

plt.title('Simple Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.grid()
plt.show()
```

```python
plt.boxplot(data, tick_labels=['Data'])

plt.title('Simple Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.grid()
plt.show()
```

```python
#Notched Box Plot
plt.boxplot(data, tick_labels=['Data'], notch=True)

plt.title('Notched Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
#Pengaturan marker untuk outlier
plt.boxplot(data, 
            tick_labels=['Data'],
            showfliers=True, 
            flierprops={'markerfacecolor':'r', 'marker':'d'})

plt.title('Pengaturan Marker')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
plt.boxplot(data, 
            tick_labels=['Data'],
            showfliers=False, 
            flierprops={'markerfacecolor':'r', 'marker':'d'})

plt.title('Pengaturan Marker')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
plt.boxplot(data, 
            tick_labels=['Data'],
            showfliers=True, 
            flierprops={'markerfacecolor':'g', 'marker':'d'})

plt.title('Pengaturan Marker')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
plt.boxplot(data, 
            tick_labels=['Data'],
            showfliers=True, 
            flierprops={'markerfacecolor':'g', 'marker':'*'})

plt.title('Pengaturan Marker')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
plt.boxplot(data, 
            tick_labels=['Data'],
            showfliers=True, 
            flierprops={'markerfacecolor':'g', 'marker':'s'})

plt.title('Pengaturan Marker')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
#Horisontal Box Plot
plt.boxplot(data, tick_labels=['Data'], vert=False)

plt.title('Horisontal Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
plt.boxplot(data, tick_labels=['Data'], vert=True)

plt.title('Horisontal Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
#Violin Plot
np.random.seed(2)
data = np.random.normal(loc=100, scale=10, size=200)
data
```

```python
#Simple Violin Plot
plt.violinplot(data)

plt.title('Simple Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
plt.violinplot(data,
               showextrema=False,
               showmeans=False,
               showmedians=False,
               quantiles=None)

plt.title('Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
plt.violinplot(data,
               showextrema=True,
               showmeans=False,
               showmedians=False,
               quantiles=None)

plt.title('Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
plt.violinplot(data,
               showextrema=True,
               showmeans=False,
               showmedians=False,
               quantiles=[0.25, 0.5, 0.75])

plt.title('Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
#Horisontal Violin Plot
plt.violinplot(data, vert=False)

plt.title('Horisontal Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
plt.violinplot(data, vert=True)

plt.title('Horisontal Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
#Kasus Twin Axes 1
x = np.linspace(1, 10, 25)
x
```

```python
fig, ax1 = plt.subplots()

ax1.plot(x, np.exp(x), 'bs-', label='exp')
ax1.set_xlabel('Sumbu X')
ax1.set_ylabel('exp')

ax2 = ax1.twinx()
ax2.plot(x, np.log(x), 'ro-', label='log')
ax2.set_ylabel('log')

fig.suptitle('Contoh Twin Axes')
fig.legend(loc='lower left')

plt.show()
```

```python
#Kasus Twin Axes 2
x = np.random.randint(16, 40, size=30)
x
```

```python
def C2F(celsius=0):
    return (celsius * 1.8) + 32

def konversi_sumbu(ax1): 
    y1, y2 = ax1.get_ylim() 
    ax2.set_ylim(C2F(y1), C2F(y2)) 
    ax2.figure.canvas.draw() 

fig, ax1 = plt.subplots() 
ax2 = ax1.twinx() 

ax1.callbacks.connect('ylim_changed', konversi_sumbu) 
ax1.plot(x) 

ax1.set_xlabel('Hari')
ax1.set_ylabel('Celsius') 
ax2.set_ylabel('Fahrenheit') 

fig.suptitle('Temperatur Udara') 
plt.show() 
```
